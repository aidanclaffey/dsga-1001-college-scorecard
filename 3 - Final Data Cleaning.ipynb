```python
import pickle
import numpy as np
import pandas as pd
```

```python
with open('clean_data.pickle','rb') as f:
    d = pickle.load(f)
    X = d['X']
    y = d['y']
    del(d)
```

```python
X.isna()
```

```python
non_empty_cols = (X.isna().sum() / X.shape[0]) != 1
non_empty_col_names = [col for col, v in non_empty_cols.to_dict().items() if v]
```

```python
X = X[non_empty_col_names]
```

```python
len(X.columns)
```

```python
col_pct = (X.isna().sum() / X.shape[0])
```

```python
col_pct_nonzero = pd.Series({c: p for c, p in col_pct.to_dict().items() if p > 0})
```

```python
col_pct_nonzero.sort_values(ascending=False)
```

```python
naive_impute_avg = [
    "PFTFAC",
    "INEXPFTE",
    "AVGFACSAL",
    "GRAD_DEBT_MDN",
    "C150_4",
    "ADM_RATE",
]
for col in naive_impute_avg:
    X[f"{col}_ISNA"] = X[col].isna().astype(int)
    X[col].fillna(X[col].mean(), inplace=True)
```
